```python
# Step 2: Upload Resume File (PDF or Word)
def upload_resume():
    print('Upload your resume file (PDF or Word):')
    uploaded = files.upload()
    resume_path = list(uploaded.keys())[0]
    return resume_path

resume_path = upload_resume()
```

```python
# Step 3: Extract Text from Resume (For PDFs or Word files)
from PyPDF2 import PdfReader
from docx import Document

def extract_text(file_path):
    text = ''
    if file_path.endswith('.pdf'):
        with open(file_path, 'rb') as file:
            reader = PdfReader(file)
            for page_num in range(len(reader.pages)):
                text += reader.pages[page_num].extract_text()
    elif file_path.endswith('.docx'):
        doc = Document(file_path)
        for para in doc.paragraphs:
            text += para.text + '\n'
    return text

resume_text = extract_text(resume_path)
```

```python
# Step 6: Generate ATS Score and Suggestions
def ats_analysis(resume_text, job_description):
    prompt = f"""
    You are an ATS expert. Compare the following resume with the job description provided.
     based on the following categories:
    - Skills Match (0-100%)
    - Experience Match(0-100%)
    - Keyword Coverage(0-100%)
    - Tools & Technologies Match (0-100%)
    Provide a overall match score (0-100%) based on skills, tools, and terminology match.
    Suggest improvements if necessary, including missing tools, terminology, and ways to optimize the resume.
    Finally, generate a customized CV tailored to the job description based on the resume content.

    Resume:
    {resume_text}

    Job Description:
    {job_description}
    """

    response = model.generate_content(prompt)

    return response.text

output = ats_analysis(resume_text, job_description)
print(output)
```
